## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df
```

```python
# Print the unique values from the "Department" column in y_df dataframe
print(y_df['Department'].unique())
```

```python
# Create a list of at least 10 column names to use as X data
X_list = ['Age', 'DistanceFromHome', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'TrainingTimesLastYear', 'WorkLifeBalance']


# Create X_df using your selected columns
X_df = attrition_df[X_list]

# Show the data types for X_df
X_df.dtypes
```

```python
X=X_df
y=y_df
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=78)

```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
X_train_scaled
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder

department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data
department_train=department_encoder.fit_transform(y_train[['Department']])
department_test=department_encoder.fit_transform(y_test[['Department']])

department_train

```

```python
# Convert the Department OneHotEncoded Arrays back to data frames
department_train_df = pd.DataFrame(department_train,index=y_train.index,columns=department_encoder.get_feature_names_out(['Department']))

department_test_df = pd.DataFrame(department_test,index=y_test.index,columns=department_encoder.get_feature_names_out(['Department']))

department_train_df
```

```python
department_test_df
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data
attrition_train=attrition_encoder.fit_transform(y_train[['Attrition']])
attrition_test=attrition_encoder.fit_transform(y_test[['Attrition']])

attrition_train
```

## Create, Compile, and Train the Model

```python
# Convert the Attrition OneHotEncoded Arrays back to data frames
attrition_train_df = pd.DataFrame(attrition_train,index=y_train.index,columns=attrition_encoder.get_feature_names_out(['Attrition']))

attrition_test_df = pd.DataFrame(attrition_test,index=y_test.index,columns=attrition_encoder.get_feature_names_out(['Attrition']))

attrition_train_df
```

```python
attrition_test_df
```

```python
# Concatenate the one-hot encoded department and attrition data with the y_train and y_test
y_train_encoded=pd.concat([y_train,department_train_df,attrition_train_df],axis=1)
y_test_encoded=pd.concat([y_test,department_test_df,attrition_test_df],axis=1)

y_train_encoded
```

```python
# Find the number of columns in the X training data
num_columns = X_train.shape[1]
print(num_columns)

# Create the input layer
input_layer = layers.Input(shape=(num_columns,))

# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu', name = 'shared_layer1')(input_layer)
shared_layer2 = layers.Dense(units=128, activation='relu', name = 'shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(units=32, activation='softmax', name='department_hidden')(shared_layer2)

# Create the output layer
department_output = layers.Dense(units=3, activation='softmax', name='department_output')(department_hidden)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(units=32, activation='sigmoid', name='attrition_hidden')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(units=2, activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam', loss={'department_output': 'categorical_crossentropy',
                    'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy',
                        'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled,
          {'department_output': department_train,
           'attrition_output': attrition_train},
          epochs=100,
          batch_size=32,
          validation_split=0.2)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'attrition_output': attrition_test, 'department_output': department_test})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Attrition Predictions Accuracy: {test_results[1]}")
print(f"Department Predictions Accuracy: {test_results[2]}")
```
